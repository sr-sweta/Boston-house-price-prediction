Feature Generation-Latitude and Longitude

```python
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent="OpenStreetMap Nominatim")

def feature_generation(df):
    lat=[]
    long=[]
    a=0
    for i in df['Location']: 
        location = geolocator.geocode(i)
        try:
            lat.append(location.latitude)
            long.append(location.longitude)
            print(a)
        except:
            lat.append("NA")
            long.append("NA")
        a=a+1
    df['Latitude'] = lat
    df['Longitude'] = long
```

```python
feature_generation(banglore_dataset)
feature_generation(Mumbai_dataset)
feature_generation(Kolkata_dataset)
feature_generation(Delhi_dataset)
```

```python
def triple_plot(x, title,c):
    fig, ax = plt.subplots(3,1,figsize=(20,10),sharex=True)
    sns.distplot(x, ax=ax[0],color=c)
    ax[0].set(xlabel=None)
    ax[0].set_title('Histogram + KDE')
    sns.boxplot(x, ax=ax[1],color=c)
    ax[1].set(xlabel=None)
    ax[1].set_title('Boxplot')
    sns.violinplot(x, ax=ax[2],color=c)
    ax[2].set(xlabel=None)
    ax[2].set_title('Violin plot')
    fig.suptitle(title, fontsize=16)
    plt.tight_layout(pad=3.0)
    plt.show()
```

```python
triple_plot(banglore_dataset['Price'],'Distribution of Price(in lakhs) in Bangalore',custom_colors[0])
```

```python
def count_plot(data,title):
    df5=data[data['Resale']==0]
    df6=data[data['Resale']==1]
    fig,ax = plt.subplots(1,2,figsize=(15,10))
    ax[0] = sns.countplot(y='Location',data=df5,order=df5.Location.value_counts().index[:10],ax=ax[0])
    ax[0].set_title('No. of new properties')
    ax[1] = sns.countplot(y='Location',data=df6,order=df6.Location.value_counts().index[:10],ax=ax[1])
    ax[1].set_title('No. of old properties')
    fig.suptitle(title,fontsize=16)
    plt.tight_layout(pad=3.0)
    plt.show()
    
```

```python
count_plot(banglore_dataset,'New and Resale Properties in Bangalore')
```

```python
count_plot(Mumbai_dataset,'New and Resale Properties in Mumbai')
```

```python
count_plot(Kolkata_dataset,'New and Resale Properties in kolkata')
```

```python
count_plot(Delhi_dataset,'New and Resale Properties in Delhi')
```

```python
def cat_plot(data,title):
    sns.catplot(x='No. of Bedrooms',y='Price',data=data)
    plt.title('No.of Bedrooms Vs Price in'+title,size=16)
    plt.gcf().set_size_inches(6,8)
    plt.show
```
